```python
from ydata_profiling import ProfileReport

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv('MiNDAT.csv', index_col='LOCAL_IDENTIFIER')
target = 'CORRUCYSTIC_DENSITY'

df.head()
```

```python
display(
    pd.concat(
        [
        df.describe().transpose(),
        df.nunique().to_frame().rename(columns={0: 'nunique'}),
        df.isna().sum().to_frame().rename(columns={0: 'null_count'})],
        axis=1
    ).style.set_caption(f"Basic description and analysis - train data")
)

```

```python
cat_cols = df.select_dtypes(include=['object', 'string']).columns.tolist()

for cat in cat_cols:
    if pd.isna(df[cat]).any():
        df[cat] = df[cat].fillna("MISSING")

```

```python
df_c = df.dropna(subset=target)
```

```python
display(
    pd.concat(
        [
        df_c.describe().transpose(),
        df_c.nunique().to_frame().rename(columns={0: 'nunique'}),
        df_c.isna().sum().to_frame().rename(columns={0: 'null_count'})],
        axis=1
    ).style.set_caption(f"Basic description and analysis - train data")
)

```

```python
def combine_two_numeric_columns(df, col1, col2):
  df[col2] = df[col2].fillna(df[col1])

  return df



df_c = combine_two_numeric_columns(df_c,'+U@','A>.')
df_c.drop(columns=['+U@'], inplace=True)

df_c = combine_two_numeric_columns(df_c,'.b6nl','Kj,')
df_c.drop(columns=['.b6nl'], inplace=True)

df_c = combine_two_numeric_columns(df_c,'Jv[i','ZVf')
df_c.drop(columns=['Jv[i'], inplace=True)

df_c = combine_two_numeric_columns(df_c,'<!!','ZZw3=!t')
df_c.drop(columns=['<!!'], inplace=True)

df_c = combine_two_numeric_columns(df_c,'3I\\y','fPqsI')
df_c.drop(columns=['3I\\y'], inplace=True)

df_c = combine_two_numeric_columns(df_c,'|G}','14W$Q')
df_c.drop(columns=['|G}'], inplace=True)

df_c.head(10)



```

```python
df_c.to_csv("MiNDAT_c.csv", index=True)
```

# Probablity var and percentage var check

```python
type_schema = {
    "maT_r":"categorical",
    "F'3Ku":"categorical",
    "MINDSPIKE_VERSION":"categorical"
}

correlations = {
    "phi_k":{"calculate": True},
    "kendall": {"calculate": True},
    "cramers": {"calculate": True}
}

profile = ProfileReport(df_c, title = 'df_c profile Report', type_schema=type_schema, interactions={'targets': ['CORRUCYSTIC_DENSITY']}, correlations=correlations,explorative=True)
```
