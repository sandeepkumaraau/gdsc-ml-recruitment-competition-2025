```python
df_x = df.copy()
```

```python
def combine_two_numeric_columns(df, col1, col2):
  df[col2] = df[col2].fillna(df[col1])

  return df



df_x = combine_two_numeric_columns(df_x,'+U@','A>.')
df_x.drop(columns=['+U@'], inplace=True)

df_x = combine_two_numeric_columns(df_x,'.b6nl','Kj,')
df_x.drop(columns=['.b6nl'], inplace=True)

df_x = combine_two_numeric_columns(df_x,'Jv[i','ZVf')
df_x.drop(columns=['Jv[i'], inplace=True)

df_x = combine_two_numeric_columns(df_x,'<!!','ZZw3=!t')
df_x.drop(columns=['<!!'], inplace=True)

df_x = combine_two_numeric_columns(df_x,'3I\\y','fPqsI')
df_x.drop(columns=['3I\\y'], inplace=True)

df_x = combine_two_numeric_columns(df_x,'|G}','14W$Q')
df_x.drop(columns=['|G}'], inplace=True)

df_x.head(10)



```

```python
df_x.drop(columns=['Z~x0<k','A>.','>?64:','U"r','TSWm',"w-u:jN'qI",'PZ8','jNhEum'],inplace=True)
```

```python
cat_cols = df_x.select_dtypes(include=['object', 'string']).columns.tolist()

for cat in cat_cols:
    if pd.isna(df[cat]).any():
        df_x[cat] = df_x[cat].fillna("MISSING")
```

```python
def sanitize_columns(df):
    df = df.copy()
    new_cols = []
    seen = {}
    for c in df.columns:
        name = re.sub(r'[\[\]<>]', '_', str(c))  # replace [, ], <
        if name in seen:
            seen[name] += 1
            name = f"{name}__{seen[name]}"      # make unique if collision
        else:
            seen[name] = 0
        new_cols.append(name)
    df.columns = new_cols
    return df

df_x = sanitize_columns(df_x)
```
